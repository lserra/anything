```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('default')
```

```python
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
path = '/home/laercio.serra/Temp/'
filename = 'bikes.csv'
```

```python
bikes = pd.read_csv(path+filename, sep='\t', encoding='utf-8', parse_dates=['Date'], dayfirst=True, index_col='Date')
```

```python
bikes.head()
```

```python
bikes.dtypes
```

```python
bikes['Berri 1'].plot()
plt.show()
```

```python
berri_bikes = bikes[['Berri 1']].copy()
berri_bikes.head()
```

```python
berri_bikes.index
```

```python
berri_bikes.index.day
```

```python
berri_bikes.index.weekday
```

```python
berri_bikes.loc[:,'weekday'] = berri_bikes.index.weekday
berri_bikes.head()
```

```python
weekday_counts = berri_bikes.groupby('weekday').aggregate(sum)
weekday_counts
```

```python
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts
```

```python
weekday_counts = berri_bikes.groupby('weekday').mean()
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts
```

```python
weekday_counts.plot(kind='bar')
plt.show()
```
