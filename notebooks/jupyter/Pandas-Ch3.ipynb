```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('default')
```

```python
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
path = '/home/laercio.serra/Temp/'
filename = '311-service-requests.csv'
```

```python
complaints = pd.read_csv(path+filename)
```

```python
complaints.head()
```

```python
noise = complaints[complaints['Complaint Type'] == "Noise - Street/Sidewalk"]
noise.head()
```

```python
is_noise = complaints['Complaint Type'] == "Noise - Street/Sidewalk"
in_brooklyn = complaints['Borough'] == "BROOKLYN"
complaints[is_noise & in_brooklyn].head()
```

```python
complaints[is_noise & in_brooklyn][['Complaint Type', 'Borough', 'Created Date', 'Descriptor']][:10]
```

```python
is_noise = complaints['Complaint Type'] == "Noise - Street/Sidewalk"
noise_complaints = complaints[is_noise]
noise_complaints['Borough'].value_counts()
```

```python
noise_complaint_counts = noise_complaints['Borough'].value_counts()
complaint_counts = complaints['Borough'].value_counts()
perc = noise_complaint_counts / complaint_counts.astype(float)
perc
```

```python
perc.plot(kind='bar')
plt.title(u'% Noise by Borough')
plt.ylabel(u'% Noise')
plt.xlabel(u'Borough')
plt.show()
```
