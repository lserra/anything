```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime as dt

%matplotlib inline
```

```python
plt.style.use('default')
```

```python
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
path = '/home/laercio.serra/Temp/'
filename = 'weather_2012.csv'
```

```python
# weather = pd.read_csv(path+filename, index_col='Date/Time')
weather = pd.read_csv(path+filename)
```

```python
weather.head()
```

```python
weather.dtypes
```

```python
# FAZENDO A CONVERSAO DA COLUNA DATE/TIME DO DATAFRAME DE OBJECT TYPE PARA DATETIME
# weather['Date'] = weather['Date/Time'].apply(lambda d: dt.strptime(d, '%Y-%m-%d %H:%M:%S'))
weather['Date'] = pd.to_datetime(weather['Date/Time'])
weather.dtypes
```

```python
weather = weather.drop('Date/Time', axis=1)
weather.head()
```

```python
new_weather = weather.set_index('Date')
```

```python
new_weather.head()
```

```python
new_weather['Temp (C)'].plot(figsize=(15, 5))
plt.show()
```

```python
# The argument axis=1 to dropna means "drop columns", not rows", and how='any' means "drop the column 
# if any value is null".
new_weather = weather.dropna(axis=1, how='any')
new_weather[:5]
```

```python
temperatures = new_weather[[u'Temp (C)', u'Date']].copy()
temperatures = temperatures.set_index('Date')
temperatures.head()
```

```python
temperatures.loc[:,'Hour'] = temperatures.index.hour
temperatures[:5]
```

```python
temperatures.groupby('Hour').aggregate(np.median).plot(figsize=(15,5))
```
